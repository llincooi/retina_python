```python
import numpy as np

s = 'frzrmzlygfveulqfpdbhlqdqrrcrwdnxeuoqqeklaitgdphcspijthbsfyfvladzpbfudkklrwqaozmixrpifeffeclhbvfukbye'
dp = np.ones([len(s),len(s)])
 
def func(i,j,s,dp):
    if i==j: return 1;
    if i>j: return 0;
    
    if s[i]==s[j]:
        dp[i][j] = func(i+1,j-1,s,dp)+2
    elif func(i+1,j,s,dp)<func(i,j-1,s,dp):
        dp[i][j] = func(i,j-1,s,dp)
    else:
        dp[i][j] = func(i+1,j,s,dp)
    
    return dp[i][j]
```
