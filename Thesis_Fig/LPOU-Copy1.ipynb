```python
def angular_gradient(x):
    diffx= np.gradient(x)
    diffx[np.abs(diffx)>0.6*np.pi] -= np.pi*diffx[np.abs(diffx)>0.6*np.pi]/abs(diffx[np.abs(diffx)>0.6*np.pi])
    return diffx
```

# Make and Plot series

```python
dt = 1/60
exp_folder = 'D:\\GoogleDrive\\retina\\Chou\'s data\\20210513\\'
files = ['merge_0727_OUsmooth_Bright_UD_G4.5_5min_Q100_6.5mW_0.5Hz',
         'merge_0727_OUsmooth_Bright_UD_G4.5_5min_Q100_6.5mW_1Hz',
         'merge_0727_OUsmooth_Bright_UD_G4.5_5min_Q100_6.5mW_2Hz',
         'merge_0727_OUsmooth_Bright_UD_G4.5_5min_Q100_6.5mW_4Hz']
Inten_list = []
rstate_list = []
T_list = []
x_list = []
for file in files:
    MBExp = loadmat(exp_folder+'merge\\'+file, squeeze_me =True)
    BarPos = MBExp['bin_pos']
    T=np.arange(len(BarPos)+1)*dt
    T_list.append( T.copy() )
    x = BarPos.copy()
    x = (x-np.mean(x))/np.std(x)
    x = x.astype('float')
    x_list.append( x.copy() )
```

```python
yf_list = []
xf_list = []
zo=0
for x in x_list:
    yf = np.abs(fft(x))
    T = T_list[zo]
    yf = 2/len(T) * yf[:(int(len(T)/2))]
    xf = np.linspace(0.0, 0.5/dt, int(len(T)/2))
#     plt.bar(xf, yf, zorder = zo)
    yf_list.append(yf.copy())
    xf_list.append(xf.copy())
    zo+=1
```

```python
plt.rcParams["font.size"] = 8
fig, ax = plt.subplots(2,1)
for x in x_list:
    ax[0].plot(x[700:900])
ax[0].set_xlabel(r'$t$ (s)', fontsize = 16)
ax[0].set_ylabel(r'$\xi_I(t)$ (mM/mm$^2$)', fontsize = 16)
# ax[0].set_xlim([0+150, 2+150])
# plt.xlim([0, 1])
ax[0].legend( ('1-Hz LPOU', '2-Hz LPOU', '3.5-Hz LPOU', '5-Hz LPOU', 'OU'),
           loc='best', prop={'size': 10})
# fig.tight_layout()
# fig.set_size_inches(10, 6)

zo=0
for yf in yf_list:
#     ax[1].bar(xf[1:], yf[1:], zorder = zo, width = 2*np.mean(np.diff(xf)) )
    ax[1].plot(xf[1:], yf[1:], zorder = zo)
    zo-=1

ax[1].set_yscale('log')
# ax[1].set_xscale('log')
ax[1].set_xlabel('frequency (Hz)', fontsize = 16)
ax[1].set_ylabel('Magnitude', fontsize = 16)
ax[1].set_xlim([0, 20])
ax[1].set_ylim([yf_list[0][-1], 1])
ax[1].legend( ('1-Hz LPOU', '2-Hz LPOU', '3.5-Hz LPOU', '5-Hz LPOU', 'OU'),loc='best', prop={'size': 12})
fig.tight_layout(h_pad=1)
fig.set_size_inches(10, 6)
```
