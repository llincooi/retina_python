```python
nstep = 5000
# x = np.zeros(nstep)
b = 100
# i0 = 2500
# for i in range(nstep):
#     a = ((i - i0)/b)**2
#     x[i] = np.exp(-a)
#     #if i> i0:
#      #   x[i] = 0.0
```

```python
x = np.sin(np.arange(nstep)/100)*0.5+0.5
plt.plot(x)
```

```python
#AGC
gain =np.ones(nstep)
gain2 =np.ones(nstep)
delay = int(10/0.05)
delay = int(b/2)
B =1
y = x.copy()
y2 = x.copy()
for i in range(delay,nstep):
    gain[i] = 1/(1+B*y[i-delay]**2)
    y[i]= gain[i]*x[i]

    gain2[i] = 1/(1+B*y2[i-delay]**2)
    y2[i] = gain2[i]*2*x[i]

```

```python
plt.plot(gain)
plt.plot(gain2)
plt.plot(x,'b')
# plt.plot(y/y.max(),'r')
plt.plot(y,'r')
plt.plot(y2/2,'g')
plt.xlim([2000,3000])
print((np.argmax(x)-np.argmax(y))/b)
```
